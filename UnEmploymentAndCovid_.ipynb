```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv('/content/Unemployment in India.csv')
data.info()

# Clean column names by removing leading/trailing spaces
data.columns = data.columns.str.strip()

# Calculate the average estimated unemployment rate for each region
unemployment_by_region = data.groupby('Region')['Estimated Unemployment Rate (%)'].mean().sort_values(ascending=False)

# Create the bar plot
plt.figure(figsize=(14, 8))
sns.barplot(x=unemployment_by_region.index, y=unemployment_by_region.values, palette='viridis')
plt.xticks(rotation=90, ha='right') # Rotate labels for better readability
plt.xlabel('Region', fontsize=12)
plt.ylabel('Average Estimated Unemployment Rate (%)', fontsize=12)
plt.title('Average Estimated Unemployment Rate (%) by Region', fontsize=14)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout() # Adjust layout to prevent labels from overlapping
plt.show()
```

```python
unemployment_by_area = data.groupby('Area')['Estimated Unemployment Rate (%)'].mean().sort_values(ascending=False)

# Create the bar plot
plt.figure(figsize=(8, 5))
sns.barplot(x=unemployment_by_area.index, y=unemployment_by_area.values, palette='viridis')
plt.xticks(rotation=90, ha='right') # Rotate labels for better readability
plt.xlabel('Area', fontsize=12)
plt.ylabel('Average Estimated Unemployment Rate (%)', fontsize=12)
plt.title('Average Estimated Unemployment Rate (%) by Area', fontsize=14)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout() # Adjust layout to prevent labels from overlapping
plt.show()
```
